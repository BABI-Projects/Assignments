## Data Partition - Train & Test Data

```r

set.seed(123)
ind <- sample.split(Y = ML3$Transport, SplitRatio = 0.7)
MLTrain <- ML3[ind,]
MLTest <- ML3[!ind,]  
```

```r

table(MLTrain$Transport)
table(MLTest$Transport)
```

# CART Technique

```r
#4.1: CART Technique
library(rpart)
library(rpart.plot)
```

```r
table(MLTrain$Transport)

```

## Model1

```r
r.ctrl = rpart.control(minsplit=100, minbucket = 10, cp = 0, xval = 10)
m1 <- rpart(formula = Transport ~ ., 
            data = MLTrain, method = "class")
m1
```

```r
rpart.plot(m1, type = 3, digits = 3, fallen.leaves = T)
```

Based on the above decision tree, we can conclude that the two most important parameters are Age and Distance. As the regression tree has split the root based on the Age which is classified as less than 31.5 years or equal to or greater than 31.5 years of age. The second most decisive variable is distance. So, based on the training dataset, employees of age 31.5 years or above, travelling to a distance of 11.1 km or above prefer car over public transport.

Employees with less than 31.5 years of age, travelling a distance of less than 10.6 km prefer public transport over any other transport.

Now, based on the model created let us make the predictions on the test dataset. The confusion matrix received from the model is as follows:



```r
p1 <- predict(m1, MLTest, type = "class")
p1
```

```r
p1.score <- predict(m1, MLTest)
printcp(m1)
plotcp(m1)
```

```r
table(MLTest$Transport, predicted = p1)
```

## model2

```r

ptree<- prune(m1, cp= 0.0021 ,"CP")
printcp(ptree)
rpart.plot(ptree, type = 3, digits = 3, fallen.leaves = T)
```

```r

p2 <- predict(ptree, MLTest, type = "class")
p2
table(MLTest$Transport, predicted = p2)
```

```r
    p2.score <- predict(ptree, MLTest)
printcp(ptree)
plotcp(ptree)
```

```r
table(MLTest$Transport, predicted = p2)
```

## Discriminant Analysis

```r
fit <- lda(MLTrain$Transport~., na.action="na.omit", data = MLTrain)
fit
```

```r
library(klaR)
```

```r
pred <- predict(fit, MLTrain)
CTable <- table(MLTrain$Transport,pred$class)
confusionMatrix(CTable)
```

```r
pred1 <- predict(fit, MLTest)
CTable1 <- table(MLTest$Transport, pred1$class)
length(MLTest$Transport)
length(pred1$class)

confusionMatrix(CTable1)
```

```r
Jacknife=lda(Transport ~., data = MLTrain, CV=TRUE)
confusionJackknife=table(Original=MLTrain$Transport,Predicted=Jacknife$class)
confusionJackknife
```

```r
Jacknife.test=lda(Transport ~., data = MLTest, CV=TRUE)
confusionJackknife.test=table(Original=MLTest$Transport,Predicted=Jacknife.test$class)
confusionJackknife.test
```

Revised accuracy is of 79%, but the accuracy of predicting Two-wheeler has decreased further, so in this case as well, we will choose Model 1 – LDA.

## support vector machine

```r
library(e1071)

set.seed(123)
```

```r
fit.svm <- svm(Transport~., data=MLTrain)
fit.svm
```

```r
svm.pred <- predict(fit.svm, MLTest)
svm.perf <- table(MLTest$Transport, svm.pred, dnn= c("Actual", "Predicted"))
svm.perf
```

```r
# Tuning an RBF support vector machine 
set.seed(123)
tuned <- tune.svm(Transport~., data=MLTrain,
                  gamma=10^(-6:1),
                  cost=10^(-10:10))
tuned
```

```r
fit.svm <- svm(Transport~., data=MLTrain, gamma=.1, cost=10)
```

```r
svm.pred <- predict(fit.svm, MLTest)
svm.perf <- table(MLTest$Transport,
                  svm.pred, dnn=c("Actual", "Predicted"))
svm.perf
```

##  Logistic regression with glm()

```r
fit.logit <- glm(Transport~., data=MLTrain, family=binomial())
```

```r
summary(fit.logit)
```

```r
summary(fit.logit)
prob <- predict(fit.logit, MLTest, type="response")
logit.pred <- factor(prob > .5, levels=c(FALSE, TRUE))
logit.perf <- table(MLTest$Transport, logit.pred,
                    dnn=c("Actual", "Predicted"))
logit.perf
```

```r
varImp(fit.logit)
```

    ## Random forest

```r
library(randomForest)
set.seed(123)
fit.forest <- randomForest(Transport~., data=MLTrain,        
                           na.action=na.roughfix,
                           importance=TRUE)             
fit.forest
```

```r
forest.pred <- predict(fit.forest, MLTest)         
forest.perf <- table(MLTest$Transport, forest.pred, 
                     dnn=c("Actual", "Predicted"))
forest.perf
```

```r
model1 <- randomForest(Transport~., data = MLTrain, ntree = 10000, mtry = 3, nodesize = 7, importance = TRUE)
model1
```

```r
Prediction_Model1 <- predict(model1, MLTest, type = "class")
Prediction_Model1
```

```r
t1 <- table(predictions = Prediction_Model1, actual = MLTest$Transport)
t1
```
